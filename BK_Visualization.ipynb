```python
# Import the Pandas library
import pandas as pd
```

```python
# Name of the CSV file
file = 'Baseball_Statistics.csv'
```

```python
# Read our Data file with the pandas library
baseball_stats_df = pd.read_csv(file, encoding="ISO-8859-1")
```

```python
# Show grid
baseball_stats_df
```

```python
# Remove the columns
baseball_stats_cleaned_df = baseball_stats_df.drop(columns=["IBB", "HBP", "SH", "SF", "GIDP", "stint"])

# Remove the dates prior to 1970
baseball_stats_cleaned_df = baseball_stats_cleaned_df[baseball_stats_cleaned_df['yearID'] >= 1970]

# Remove all AB less than 100
baseball_stats_cleaned_df = baseball_stats_cleaned_df[baseball_stats_cleaned_df['AB'] >= 100]

baseball_stats_cleaned_df
```

```python
bins = [1969, 1979, 1989, 1999, 2009, 2016]
labels = ['1970-1979', '1980-1989', '1990-1999', '2000-2009', '2010-2015']
year_bins = pd.cut(baseball_stats_cleaned_df['yearID'], bins=bins, labels=labels, right=False)

baseball_stats_cleaned_df['year_bins'] = year_bins
```

```python
grouped = baseball_stats_cleaned_df.groupby('year_bins').agg({'HR': 'sum', 'R': 'sum', 'AB': 'sum'})
grouped['AB / HR'] = grouped['AB'] / grouped['HR']
grouped['AB / R'] = grouped['AB'] / grouped['R']

grouped
```

```python
analysis = baseball_stats_cleaned_df.groupby('yearID').agg({'HR': 'sum',
                                                            'R': 'sum',
                                                            'H': 'sum',
                                                            '2B': 'sum',
                                                            '3B': 'sum',
                                                            'RBI': 'sum',
                                                            'SB': 'sum',
                                                            'CS': 'sum',
                                                            'BB': 'sum',
                                                            'SO': 'sum',
                                                            'AB': 'sum'})


analysis['AB_per_R'] = analysis['AB'] / analysis['R']
analysis['AB_per_H'] = analysis['AB'] / analysis['H']
analysis['AB_per_2B'] = analysis['AB'] / analysis['2B']
analysis['AB_per_3B'] = analysis['AB'] / analysis['3B']
analysis['AB_per_HR'] = analysis['AB'] / analysis['HR']
analysis['AB_per_RBI'] = analysis['AB'] / analysis['RBI']
analysis['AB_per_SB'] = analysis['AB'] / analysis['SB']
analysis['AB_per_CS'] = analysis['AB'] / analysis['CS']
analysis['AB_per_BB'] = analysis['AB'] / analysis['BB']
analysis['AB_per_SO'] = analysis['AB'] / analysis['SO']


analysis = analysis.drop(columns=['AB','HR', 'R', 'H', '2B', '3B', 'RBI', 'SB', 'CS', 'BB', 'SO'])

analysis
```

```python
import numpy as np

analysis2 = baseball_stats_cleaned_df.groupby('playerID').agg({'HR': 'sum',
                                                            'R': 'sum',
                                                            'H': 'sum',
                                                            '2B': 'sum',
                                                            '3B': 'sum',
                                                            'RBI': 'sum',
                                                            'SB': 'sum',
                                                            'CS': 'sum',
                                                            'BB': 'sum',
                                                            'SO': 'sum',
                                                            'AB': 'sum'})

analysis2['AB_per_R'] = analysis2['AB'] / analysis2['R']
analysis2['AB_per_H'] = analysis2['AB'] / analysis2['H']
analysis2['AB_per_2B'] = analysis2['AB'] / analysis2['2B']
analysis2['AB_per_3B'] = analysis2['AB'] / analysis2['3B']
analysis2['AB_per_HR'] = analysis2['AB'] / analysis2['HR']
analysis2['AB_per_RBI'] = analysis2['AB'] / analysis2['RBI']
analysis2['AB_per_SB'] = analysis2['AB'] / analysis2['SB']
analysis2['AB_per_CS'] = analysis2['AB'] / analysis2['CS']
analysis2['AB_per_BB'] = analysis2['AB'] / analysis2['BB']
analysis2['AB_per_SO'] = analysis2['AB'] / analysis2['SO']

analysis2 = analysis2.drop(columns=['AB','HR', 'R', 'H', '2B', '3B', 'RBI', 'SB', 'CS', 'BB', 'SO'])

analysis2 = analysis2.replace([np.inf, -np.inf], np.nan)
analysis2 = analysis2.dropna()

analysis2
```

```python
analysis3 = baseball_stats_cleaned_df.groupby('playerID').agg({'R': 'sum',
                                                                'H': 'sum',
                                                                '2B': 'sum',
                                                                '3B': 'sum',
                                                                'HR': 'sum',
                                                                'RBI': 'sum',
                                                                'SB': 'sum',
                                                                'CS': 'sum',
                                                                'BB': 'sum',
                                                                'SO': 'sum'})

remove_zero = ['R', 'H', '2B', '3B', 'HR', 'RBI', 'SB', 'CS', 'BB', 'SO']
analysis3 = analysis3[(analysis3[remove_zero] != 0).all(axis=1)]

analysis3.head()
```
