```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st

```

```python
csv_path = "../Project_1/Baseball_Statistics_2.csv"
baseball_stats_df = pd.read_csv(csv_path)

baseball_stats_df.head()


```

```python
baseball_stats_cleaned_df = baseball_stats_df.drop(columns=["IBB", "HBP", "SH", "SF", "GIDP", "stint"]) 
baseball_stats_cleaned_df = baseball_stats_cleaned_df[baseball_stats_cleaned_df['yearID'] >=1970]
baseball_stats_cleaned_df = baseball_stats_cleaned_df[baseball_stats_cleaned_df['yearID'] <=1976]
baseball_stats_cleaned_df = baseball_stats_cleaned_df[baseball_stats_cleaned_df['AB'] >= 100]
baseball_stats_cleaned_df
```

```python
aaronha01_stats_df = baseball_stats_cleaned_df[baseball_stats_cleaned_df["playerID"]== "aaronha01"]
```

```python
aaronha01_stats_df 
```

```python
aaronha01_stats_df = baseball_stats_cleaned_df[baseball_stats_cleaned_df["playerID"]== "aaronha01"]
grouped_df = aaronha01_stats_df.groupby('yearID').agg({'HR': 'sum', 'AB': 'sum'})
grouped_df['HR / AB'] = grouped_df['HR'] / grouped_df['AB']

grouped_df
```

```python
baseball_stats_cleaned_df = baseball_stats_cleaned_df[baseball_stats_cleaned_df['yearID'] >= 1970]
baseball_stats_cleaned_df


                                           
```

```python
baseball_stats_cleaned_df = baseball_stats_cleaned_df[baseball_stats_cleaned_df['yearID'] >=1970]
baseball_stats_cleaned_df = baseball_stats_cleaned_df[baseball_stats_cleaned_df['yearID'] <=1976]
analysis = baseball_stats_cleaned_df.groupby('yearID').agg({'HR': 'sum','AB': 'sum'})



analysis['HR / AB'] = analysis['HR'] / analysis['AB']
analysis



```

```python
import matplotlib.pyplot as plt



plt.figure(figsize=(12,8))
plt.plot(grouped_df.index, grouped_df['HR / AB'], label='HR / AB')
plt.plot(analysis.index, analysis['HR / AB'], label='HR / AB')


plt.xlabel('Year')
plt.ylabel('HR / AB')
plt.title('HR / AB')
plt.legend(loc='best')


plt.show()
```
