```python
# Import the Pandas library
import pandas as pd
```

```python
# Name of the CSV file
file = 'Baseball_Statistics.csv'
```

```python
# Read our Data file with the pandas library
baseball_stats_df = pd.read_csv(file, encoding="ISO-8859-1")
```

```python
# Show grid
baseball_stats_df
```

```python
# Remove the columns
baseball_stats_cleaned_df = baseball_stats_df.drop(columns=["IBB", "HBP", "SH", "SF", "GIDP", "stint"])

# Remove the dates prior to 1970
baseball_stats_cleaned_df = baseball_stats_cleaned_df[baseball_stats_cleaned_df['yearID'] >= 1970]

# Remove all AB less than 100
baseball_stats_cleaned_df = baseball_stats_cleaned_df[baseball_stats_cleaned_df['AB'] >= 100]

baseball_stats_cleaned_df
```

```python
bins = [1969, 1979, 1989, 1999, 2009, 2016]
labels = ['1970-1979', '1980-1989', '1990-1999', '2000-2009', '2010-2015']
year_bins = pd.cut(baseball_stats_cleaned_df['yearID'], bins=bins, labels=labels, right=False)

baseball_stats_cleaned_df['year_bins'] = year_bins
```

```python
grouped = baseball_stats_cleaned_df.groupby('year_bins').agg({'SO': 'sum', 'AB': 'sum'})
grouped['SO / AB'] = grouped['SO'] / grouped['AB']

grouped
```

```python
analysis = baseball_stats_cleaned_df.groupby('yearID').agg({'HR': 'sum',
                                                            'R': 'sum',
                                                            'H': 'sum',
                                                            '2B': 'sum',
                                                            '3B': 'sum',
                                                            'RBI': 'sum',
                                                            'SB': 'sum',
                                                            'CS': 'sum',
                                                            'BB': 'sum',
                                                            'SO': 'sum',
                                                            'AB': 'sum'})


analysis['R / AB'] = analysis['R'] / analysis['AB']
analysis['H / AB'] = analysis['H'] / analysis['AB']
analysis['2B / AB'] = analysis['2B'] / analysis['AB']
analysis['3B / AB'] = analysis['3B'] / analysis['AB']
analysis['HR / AB'] = analysis['HR'] / analysis['AB']
analysis['RBI / AB'] = analysis['RBI'] / analysis['AB']
analysis['SB / AB'] = analysis['SB'] / analysis['AB']
analysis['CS / AB'] = analysis['CS'] / analysis['AB']
analysis['BB / AB'] = analysis['BB'] / analysis['AB']
analysis['SO / AB'] = analysis['SO'] / analysis['AB']

analysis = analysis.drop(columns=['AB','HR', 'R', 'H', '2B', '3B', 'RBI', 'SB', 'CS', 'BB', 'SO'])

analysis
```

```python
import numpy as np

analysis2 = baseball_stats_cleaned_df.groupby('playerID').agg({'HR': 'sum',
                                                            'R': 'sum',
                                                            'H': 'sum',
                                                            '2B': 'sum',
                                                            '3B': 'sum',
                                                            'RBI': 'sum',
                                                            'SB': 'sum',
                                                            'CS': 'sum',
                                                            'BB': 'sum',
                                                            'SO': 'sum',
                                                            'AB': 'sum'})

analysis2['AB / R'] = analysis2['AB'] / analysis2['R']
analysis2['AB / H'] = analysis2['AB'] / analysis2['H']
analysis2['AB / 2B'] = analysis2['AB'] / analysis2['2B']
analysis2['AB / 3B'] = analysis2['AB'] / analysis2['3B']
analysis2['AB / HR'] = analysis2['AB'] / analysis2['HR']
analysis2['AB / RBI'] = analysis2['AB'] / analysis2['RBI']
analysis2['AB / SB'] = analysis2['AB'] / analysis2['SB']
analysis2['AB / CS'] = analysis2['AB'] / analysis2['CS']
analysis2['AB / BB'] = analysis2['AB'] / analysis2['BB']
analysis2['AB / SO'] = analysis2['AB'] / analysis2['SO']

analysis2 = analysis2.drop(columns=['AB','HR', 'R', 'H', '2B', '3B', 'RBI', 'SB', 'CS', 'BB', 'SO'])

analysis2 = analysis2.replace([np.inf, -np.inf], np.nan)
analysis2 = analysis2.dropna()

analysis2
```

```python
analysis3 = baseball_stats_cleaned_df.groupby('playerID').agg({'R': 'sum',
                                                                'H': 'sum',
                                                                '2B': 'sum',
                                                                '3B': 'sum',
                                                                'HR': 'sum',
                                                                'RBI': 'sum',
                                                                'SB': 'sum',
                                                                'CS': 'sum',
                                                                'BB': 'sum',
                                                                'SO': 'sum'})

remove_zero = ['R', 'H', '2B', '3B', 'HR', 'RBI', 'SB', 'CS', 'BB', 'SO']
analysis3 = analysis3[(analysis3[remove_zero] != 0).all(axis=1)]

analysis3.head()
```
